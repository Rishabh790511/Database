```python
#GET Method: It is a request method used to request data from a specified resource. It is one of the HTTP methods that clients can use when making requests to a server. Data sent using GET method is visible to everyone in the URL, and it should only be used when the data sent to the server is not sensitive.

POST Method: It is a request method used when submitting data to be processed to a specified resource. It is one of the HTTP methods that clients can use to send data to the server to create/update a resource. Data sent using POST method is not visible in the URL, and it is more secure than the GET method.
```

```python
#Q4
Templates in Flask are used to separate the presentation layer from the business logic. They allow you to dynamically generate HTML content by embedding variables, control structures, and other dynamic elements. The render_template() function is used to render these templates and return the resulting HTML to the client. It makes it easier to create dynamic web pages by combining static HTML with dynamic content generated by the Flask application.
```

```python
#Q5
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route('/api/greet', methods=['POST'])
def greet():
    data = request.get_json()
    name = data.get('name', '')
    greeting = f"Hello, {name}!"
    return jsonify({'greeting': greeting})

if __name__ == '__main__':
    app.run(debug=True)
```
